# Using GPkit for Aircraft Optimization
-----------------------------------

Based on "Geometric Programming for Aircraft Design Optimization" by Hoburg and Abbeel, and using [GPkit](https://github.com/appliedopt/gpkit) for model representation and solver interfacing.

## Objective

We want to make the airplane use as little fuel as possible when it's cruising, while keeping in mind:

  - the flow's Reynolds number $Re$
  - the skin friction coefficient $C_f$ (for which we assume a fully turbulent boundary layer)
  - the fuselage's viscous drag, the wing's parasitic and induced drag
  - that at cruising speeds the plane weighs less than the lift it generates, i.e. flies
  - that at takeoff speeds the plane weighs less than the lift it generates, i.e. flies
  - that the plane weighs as much as the wing plus everything else
  - that the wing's weight depends on its design and on the weight it has to lift
  
Or in equation form:

$$\begin{array}{ll}
\mbox{minimize}
    & \frac{1}{2} \rho S C_D V^2\\
\mbox{subject to}
    & Re \leq \frac{\rho V}{\mu}\sqrt{\frac{S}{A}} \\
    & C_f \geq \frac{0.074}{Re^{0.2}} \\
    & C_D \geq \frac{(CDA_0)}{S} + k C_f \frac{S_{wet}}{S} + \frac{C_L^2}{\pi A e} \\
    & W \leq \frac{1}{2} \rho S C_L V^2\\
    & W \leq \frac{1}{2} \rho S C_{L,max} V_{min}^2\\
    & W \geq W_0 + W_w \\
    & W_w \geq 45.42 S + 8.71\times10^{-5} \frac{N_{ult} A^{3/2} \sqrt{W_0 W S}}{\tau} \\
 \end{array}$$
 
 We'll solve this for cruising speeds between 45 and 55 m/s, and takeoff speeds between 20 and 25 m/s, and get the following plots:

```python
import numpy as np
import gpkit
gpkit.init_printing()
```

```python
constants = {
    'pi': np.pi,
    'CDA_0': (0.031, "[m^2] fuselage drag area"),
    'rho': (1.23, "[kg/m^3] density of air"),
    'mu': (1.78e-5, "[kg/(m*s)] viscosity of air"),
    'S_wetratio': (2.05, "[-] wetted area ratio"),
    'k': (1.2, "form factor"),
    'e': (0.95, "Oswald efficiency factor"),
    'W_0': (4940, "[N] aircraft weight excluding wing"),
    'N_ult': (3.8, "ultimate load factor"),
    'tau': (0.12, "airfoil thickness to chord ratio"),
    'C_Lmax': (1.5, "max CL, flaps down"),
    'V': ('sweep', np.linspace(45, 55, 10), "[m/s] cruising speed"),
    'V_min': ('sweep', np.linspace(20, 25, 10), "[m/s] takeoff speed"),
}
gpkit.monify_up(globals(), constants)
```

```python
free_variables = {
  'A': "aspect ratio",
  'S': "[m^2] total wing area",
  'C_D': "Drag coefficient of wing",
  'C_L': "Lift coefficent of wing",
  'C_f': "skin friction coefficient",
  'Re': "Reynold's number",
  'W': "[N] total aircraft weight",
  'W_w': "[N] wing weight",
}
gpkit.monify_up(globals(), free_variables)
```

```python
Re = (rho/mu)*V*(S/A)**0.5
C_f = 0.074/Re**0.2
D = 0.5*rho*S*C_D*V**2
```

```python
C_f
```

```python
# drag modeling #
C_D_fuse = CDA_0/S            # fuselage viscous drag
C_D_wpar = k*C_f*S_wetratio # wing parasitic drag
C_D_ind = C_L**2/(pi*A*e)    # induced drag

# wing-weight modeling #
W_w_surf = 45.24*S                                   # surface weight
W_w_strc = 8.71e-5*(N_ult*A**1.5*(W_0*W*S)**0.5)/tau # structural weight

gp = gpkit.GP( # minimize
               D, 
             [ # subject to
              C_D >= C_D_fuse + C_D_wpar + C_D_ind,
              W <= 0.5*rho*S*C_L*V**2,
              W <= 0.5*rho*S*C_Lmax*V_min**2,
              W >= W_0 + W_w,
              W_w >= W_w_surf + W_w_strc,
             ], constants=constants)
```

```python
data = gp.solve()
```
