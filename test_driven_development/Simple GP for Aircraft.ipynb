## Objective

The following optimization is represented in the code below:

$$\begin{array}{ll}
\mbox{minimize}
    & \frac{1}{2} \rho V^2 C_D S\\
\mbox{subject to}
    & Re \leq \frac{\rho V}{\mu}\sqrt{\frac{S}{A}} \\
    & C_f \geq \frac{0.074}{Re^{0.2}} \\
    & C_D \geq \frac{(CDA_0)}{S} + k C_f \frac{S_{wet}}{S} + \frac{C_L^2}{\pi A e} \\
    & \frac{1}{2} \rho V^2 C_L S \geq W \\
    & W \geq W_0 + W_w \\
    & W_w \geq 45.42 S + 8.71\times10^{-5} \frac{N_{ult} A^{3/2} \sqrt{W_0 W S}}{\tau} \\
    & C_{L,max} \geq \frac{2 W}{\rho V^2_{min} S}
 \end{array}$$

```python
from numpy import linspace, zeros, pi
import gpkit
```

```python
CDA0 = 0.03062702    # [m^2] fuselage drag area
rho = 1.23           # [kg/m^3] density of air
mu = 1.78e-5         # [kg/(m*s)] viscosity of air
S_wet_ratio = 2.05   # [-] wetted area ratio
k = 1.2              # [-] form factor
e = 0.96             # [-] Oswald efficiency factor
W_0 = 4940           # [N] aircraft weight excluding wing
N_ult = 2.5          # [-] ultimate load factor
tau = 0.12           # [-] airfoil's thickness-to-chord ratio
C_Lmax = 2.0         # [-] max CL with the flaps down (takeoff)
```

```python
varstring = 'A S  C_D  C_L  C_f  Re  W  W_w'
(A,   # [-] aspect ratio
S,    # [m^2] total area
C_D,  # [-] Drag coefficient of wing
C_L,  # [-] Lift coefficent of wing
C_f,  # [-] skin friction coefficient
Re,   # [-] Reynold's number
W,    # [N] total aircraft weight
W_w   # [N] wing weight
) = gpkit.monify(varstring)

# drag model #
C_D_fuse = CDA0/S             # [-] fuselage viscous drag
C_D_wpar = k*C_f*S_wet_ratio  # [-] wing parasitic drag
C_D_ind = C_L**2/(pi*A*e)     # [-] induced drag

# wing-weight model #
W_w_surf = 45.24*S                                    # [N] surface weight
W_w_strc = 8.71e-5*(N_ult*A**1.5*(W_0*W*S)**0.5)/tau  # [N] structural weight
```

```python
resolution = (30,30)                      # [-] number of points to sweep within
V_range = linspace(45, 55, shape[0])      # [m/s] cruising speed and
V_min_range = linspace(20, 25, shape[1])  # [m/s] takeoff speed
```

```python
%%time

data = {var: zeros(resolution) for var in varstring.split()}

for i, V in enumerate(V_range):
  for j, V_min in enumerate(V_min_range):
    s = gpkit.minimize(                            # What's the lowest
            0.5*rho*S*C_D*V**2,                    # [N] TOTAL DRAG FORCE
           [ # subject to                          # That we can get, with our
            Re <= (rho/mu)*V*(S/A)**0.5,           # flow characteristics,
            C_f >= 0.074/Re**0.2,                  # turbulent BL approximation,
            C_D >= C_D_fuse + C_D_wpar + C_D_ind,  # the above 'drag model',
            W <= 0.5*rho*S*C_L*V**2,               # flight at cruising,
            W <= 0.5*rho*S*C_Lmax*V_min**2,        # flight at takeoff,     
            W >= W_0 + W_w,                        # the plane's weight, and
            W_w >= W_w_surf + W_w_strc             # the above 'wing-weight model'?
           ],
            'cvxopt', {'show_progress': False})
    
    for var in s: data[var][i,j] = s[var]
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

def contour_plot(title, Z, contour_format='%2.1f', colors='#80cdc1', '#888888', '#01665e'):
  fig = plt.figure(figsize=(10,5))
  ax = fig.gca()

  contf = ax.contour(V_range, V_min_range, Z.T, 48, linewidths = 0.5, colors=light)
  cont = ax.contour(V_range, V_min_range, Z.T, 8, linewidths = 1, colors=dark)
  
  light, neutral, dark = colors
  ax.set_xlabel('cruise vel V (m/s)', color=neutral)
  ax.set_xlim(45, 55)
  ax.set_xticks(linspace(45,55,11),minor=True)
  ax.set_ylabel('landing vel V_min (m/s)', color=neutral)
  ax.set_ylim(20, 25)
  plt.clabel(cont, fmt = contour_format, colors=dark, fontsize=14)
  ax.set_title(title, color=dark, fontsize=14)
  ax.tick_params(colors=neutral)
  ax.set_frame_on(False)
  plt.show()
```
