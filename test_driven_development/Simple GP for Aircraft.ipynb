# Using GPkit for Aircraft Optimization
-----------------------------------

Based on "Geometric Programming for Aircraft Design Optimization" by Hoburg and Abbeel, and using [GPkit](https://github.com/appliedopt/gpkit) for model representation and solver interfacing.

## Objective

We want to make the airplane use as little fuel as possible when it's cruising, while keeping in mind:

  - the flow's Reynolds number $Re$
  - the skin friction coefficient $C_f$ (for which we assume a fully turbulent boundary layer)
  - the fuselage's viscous drag, the wing's parasitic and induced drag
  - that at cruising speeds the plane weighs less than the lift it generates, i.e. flies
  - that at takeoff speeds the plane weighs less than the lift it generates, i.e. flies
  - that the plane weighs as much as the wing plus everything else
  - that the wing's weight depends on its design and on the weight it has to lift
  
Or in equation form:

$$\begin{array}{ll}
\mbox{minimize}
    & \frac{1}{2} \rho S C_D V^2\\
\mbox{subject to}
    & Re \leq \frac{\rho V}{\mu}\sqrt{\frac{S}{A}} \\
    & C_f \geq \frac{0.074}{Re^{0.2}} \\
    & C_D \geq \frac{(CDA_0)}{S} + k C_f \frac{S_{wet}}{S} + \frac{C_L^2}{\pi A e} \\
    & W \leq \frac{1}{2} \rho S C_L V^2\\
    & W \leq \frac{1}{2} \rho S C_{L,max} V_{min}^2\\
    & W \geq W_0 + W_w \\
    & W_w \geq 45.42 S + 8.71\times10^{-5} \frac{N_{ult} A^{3/2} \sqrt{W_0 W S}}{\tau} \\
 \end{array}$$
 
 We'll solve this for cruising speeds between 45 and 55 m/s, and takeoff speeds between 20 and 25 m/s, and get the following plots:

```python
from numpy import linspace, zeros, pi
import gpkit
```

```python
from numpy import pi

constants = {
  'CDA0': 0.03062702,  # [m^2] fuselage drag area
  'rho': 1.23,         # [kg/m^3] density of air
  'mu': 1.78e-5,       # [kg/(m*s)] viscosity of air
  'S_wet_ratio': 2.05, # [-] wetted area ratio
  'k': 1.2,            # [-] form factor
  'e': 0.96,           # [-] Oswald efficiency factor
  'W_0': 4940,         # [N] aircraft weight excluding wing
  'N_ult': 2.5,        # [-] ultimate load factor
  'tau': 0.12,         # [-] airfoil thickness to chord ratio
  'C_Lmax': 2.0,       # [-] max CL, flaps down
}
gpkit.monify_up(globals(), constants)
```

```python
freevars = [
  'A',     # [-] aspect ratio
  'S',     # [m^2] total wing area
  'C_D',   # [-] Drag coefficient of wing
  'C_L',   # [-] Lift coefficent of wing
  'C_f',   # [-] skin friction coefficient
  'Re',    # [-] Reynold's number
  'W',     # [N] total aircraft weight
  'W_w'    # [N] wing weight
  'V',     # [m/s] cruising speed, to be iterated over
  'V_min', # [m/s] takeoff speed, to be iterated over 
]
gpkit.monify_up(globals(), freevars)
```

```python
# drag modeling #
C_D_fuse = CDA0/S            # fuselage viscous drag
C_D_wpar = k*C_f*S_wet_ratio # wing parasitic drag
C_D_ind = C_L**2/(pi*A*e)    # induced drag

# wing-weight modeling #
W_w_surf = 45.24*S                                   # surface weight
W_w_strc = 8.71e-5*(N_ult*A**1.5*(W_0*W*S)**0.5)/tau # structural weight

gp = gpkit.GP( # minimize
               0.5*rho*S*C_D*V**2, 
             [ # subject to
              Re <= (rho/mu)*V*(S/A)**0.5,
              C_f >= 0.074/Re**0.2,
              W <= 0.5*rho*S*C_L*V**2,
              W <= 0.5*rho*S*C_Lmax*V_min**2,
              W >= W_0 + W_w,
              W_w >= W_w_surf + W_w_strc,
              C_D >= C_D_fuse + C_D_wpar + C_D_ind
             ], solver='mosek')
```

```python
%%time

from numpy import linspace
from numpy import zeros

shape = (30,30)
Vs = linspace(45, 55, shape[0])
V_mins = linspace(20, 25, shape[1])
data = {var: zeros(shape) for var in freevars}

for i, V_ in enumerate(Vs):
    constants.update({'V': V_})
    for j, V_min_ in enumerate(V_mins):
        constants.update({'V_min': V_min_})
        gp.replace_constants(constants)
        sol = gp.solve()
        for var in sol:  data[var][i,j] = sol[var]
```

```python
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
import matplotlib.pyplot as plt

def contour_plot(ax, title, Z, contour_format='%2.1f', colors=['#80cdc1', '#888888', '#01665e']):
  light, neutral, dark = colors
  contf = ax.contour(Vs, V_mins, Z.T, 64, linewidths = 0.5, colors=light)
  cont = ax.contour(Vs, V_mins, Z.T, 4, linewidths = 1, colors=dark)
  plt.clabel(cont, fmt = contour_format, colors=dark, fontsize=14)
  ax.set_title(title, color=dark, fontsize=14)
  ax.tick_params(colors=neutral)
  ax.set_frame_on(False)
```

```python
from numpy import array, matrix, ones
F_D = 0.5*constants['rho']*data['S']*data['C_D'] * array(matrix(Vs).T*ones((1,30)))**2

fig, axes = plt.subplots(3,2, sharex=True, sharey=True, figsize=(10,15))

neutral = '#888888'
for ax in axes[:,0]:
    ax.set_ylabel('landing vel V_min (m/s)', color=neutral)
    ax.set_ylim(20, 25)
for ax in axes[-1,:]:
    ax.set_xlabel('cruise vel V (m/s)', color=neutral)
    ax.set_xlim(45, 55)
    ax.set_xticks(linspace(45,55,11))
fig.tight_layout(h_pad=3)

contour_plot(axes[0,0], 'Cruising drag, D [N]', F_D, '%2.0f')
contour_plot(axes[0,1], 'Drag coefficient, C_D [-]', data['C_D'], '%.3f')
contour_plot(axes[1,0], 'Wing weight, W_w [N]', data['W_w'], '%2.0f')
contour_plot(axes[1,1], 'Lift coefficient, C_L [-]', data['C_L'], '%.2f')
contour_plot(axes[2,0], 'Wing area, S [m^2]', data['S'])
contour_plot(axes[2,1], 'Aspect ratio, A [-]', data['A'])
```
